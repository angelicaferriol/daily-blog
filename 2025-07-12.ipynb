# Daily Blog #73 - Sets, Strings, and Languages
### July 12, 2025 


## 1. **Alphabet (Σ)**

An **alphabet** is a finite set of symbols.

> Example:
> Σ = {0, 1} → binary alphabet
> Σ = {a, b, c} → alphabet of letters
> Σ = {↑, ↓, ←, →} → arrows for robot instructions


## 2. **String**

A **string** is a finite sequence of symbols from an alphabet.

> Example:
> Σ = {0,1}
> 0101, 0011, and 1 are all strings over Σ
> "" (empty string) is called **epsilon** (ε)


## 3. **Set of Strings (Language)**

A **language** is any set of strings made from an alphabet.

> Example:
> L = {“a”, “aa”, “aaa”, …} — strings of a’s
> L = {w | w contains equal number of 0s and 1s}
> L = all palindromes over Σ = {0,1}

A language can be **finite** or **infinite**.


## 4. **Basic Operations on Strings**

Given strings `x = "ab"` and `y = "cd"`:

| Operation            | Result   |     |     |
| -------------------- | -------- | --- | --- |
| Concatenation (`xy`) | `"abcd"` |     |     |
| Length (\`           | x        | \`) | `2` |
| Reverse (`x^R`)      | `"ba"`   |     |     |


## 5. **Basic Operations on Languages**

Let’s say:
A = {“a”, “aa”}
B = {“b”, “bb”}

| Operation             | Result                       |
| --------------------- | ---------------------------- |
| **Union** A ∪ B       | {“a”, “aa”, “b”, “bb”}       |
| **Concatenation** A·B | {“ab”, “abb”, “aab”, “aabb”} |
| **Kleene Star** A\*   | {ε, a, aa, aaa, …}           |

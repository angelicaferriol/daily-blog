## I. FOUNDATIONS OF REMOTE SENSING

### What is Remote Sensing?

Remote sensing is the science of obtaining information about objects or areas from a distance, typically using aircraft or satellites.

> "Measuring without touching."

### Two Main Types:

1. **Passive Remote Sensing**

   * Detects natural radiation (e.g., sunlight) reflected or emitted.
   * Example: Optical sensors, thermal infrared.

2. **Active Remote Sensing**

   * Emits its own signal and measures the return.
   * Example: RADAR, LiDAR.

## II. SATELLITE SYSTEMS

### Satellite Orbits

* **Geostationary (GEO)**: Always above the same spot; good for weather monitoring.
* **Polar-Orbiting (LEO)**: Circles from pole to pole; better for high-resolution Earth observation.
* **Sun-synchronous Orbit**: Passes over the same part of Earth at the same solar time every day.

### Examples of Satellite Programs:

| Program     | Operator    | Sensors   | Applications                          |
| ----------- | ----------- | --------- | ------------------------------------- |
| Landsat     | NASA/USGS   | OLI, TIRS | Land use, vegetation, water quality   |
| Sentinel    | ESA         | MSI, SAR  | Environmental monitoring, agriculture |
| MODIS       | NASA        | MODIS     | Ocean color, land cover               |
| PlanetScope | Planet Labs | RGB/NIR   | Daily Earth imagery                   |

## III. REMOTE SENSING COMPONENTS

### 1. **Sensors**

Sensors capture electromagnetic radiation (EMR). They differ in:

* **Spectral resolution**: Ability to distinguish wavelengths (bands)
* **Spatial resolution**: Size of the pixel (e.g., 10m, 30m)
* **Temporal resolution**: Revisit frequency
* **Radiometric resolution**: Bit depth (e.g., 8-bit, 16-bit)

### 2. **Electromagnetic Spectrum**

Key ranges:

* **Visible** (400–700 nm)
* **Near Infrared (NIR)** (700–1400 nm)
* **Shortwave Infrared (SWIR)** (1400–3000 nm)
* **Thermal Infrared (TIR)** (3–14 µm)
* **Microwave** (for RADAR)

## IV. IMAGE INTERPRETATION

### Visual Interpretation Elements:

1. **Tone/Color** – brightness or color in a band or RGB composite
2. **Shape** – man-made vs. natural
3. **Size** – relative and absolute
4. **Pattern** – geometric vs. random
5. **Texture** – smooth vs. rough (fine vs. coarse)
6. **Shadow** – helps estimate height
7. **Association** – context clues (e.g., roads next to buildings)

### Spectral Signatures

Each material (vegetation, water, soil, etc.) reflects EMR differently across bands.

| Feature            | NIR Reflection | Visible Reflection |
| ------------------ | -------------- | ------------------ |
| Healthy Vegetation | High           | Low                |
| Water              | Low            | Low                |
| Bare Soil          | Medium         | Medium             |

## V. IMAGE PROCESSING TECHNIQUES

### Preprocessing

* **Radiometric Correction**: Fix sensor noise or atmospheric effects
* **Geometric Correction**: Align to real-world map coordinates
* **Cloud Masking**: Remove cloud-covered pixels

### Enhancement

* **Contrast Stretching**
* **Band Combinations (e.g., RGB = NIR, Red, Green)** for false-color images
* **Histogram Equalization**

### Index Calculation

Common indices (used for classification):

| Index             | Formula                       | Use               |
| ----------------- | ----------------------------- | ----------------- |
| NDVI (Vegetation) | (NIR - Red) / (NIR + Red)     | Vegetation health |
| NDWI (Water)      | (Green - NIR) / (Green + NIR) | Water detection   |
| NDBI (Built-up)   | (SWIR - NIR) / (SWIR + NIR)   | Urban areas       |

## VI. CLASSIFICATION METHODS

### 1. **Unsupervised Classification**

* Algorithm groups similar pixels into clusters (e.g., K-means).
* No prior training data needed.
* Use case: initial exploratory analysis.

### 2. **Supervised Classification**

* Requires **training data**.
* Classifies based on labeled regions.
* Algorithms: Maximum Likelihood, SVM, Random Forest, etc.

### 3. **Object-Based Image Analysis (OBIA)**

* Groups pixels into objects (segments).
* Better for high-resolution data and urban features.
